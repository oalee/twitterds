```python

import yerbamate
env = yerbamate.Environment()

from twitter.data.loader.user import get_tweets_user

user_name = "DaiKatuo"

df = get_tweets_user(user_name)
# Group by language and count the number of tweets
language_df = df.groupby("lang").size().reset_index(name="count")

# Plot the language counts
language_df.plot(x="lang", y="count", kind="bar", title=f"Language Distribution of {user_name}'s Tweets")

# Print the number of unique languages
print(f"Number of unique languages: {language_df.shape[0]}")
```
