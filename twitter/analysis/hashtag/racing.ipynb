```python
import yerbamate, pandas as pd, os
env = yerbamate.Environment()

hashtags_path = os.path.join(env["plots"], "analysis", "hashtag.parquet")

df = pd.read_parquet(hashtags_path)
df.head()
```

```python
font_path ='/usr/share/fonts/TTF/XB-Zar-Regular.ttf'


# Convert year, month, and day columns to datetime
df["date"] = pd.to_datetime(df[["year", "month", "day"]])

# Split the dataframe into two based on the date
df = df[df["date"] < "2022-04-23"]

df_pivot = df.pivot(index='date', columns='hashtag', values='count')
# df_pivot = df_pivot.fillna(0)
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
from matplotlib.ticker import EngFormatter
from matplotlib.font_manager import FontProperties

font_path = '/usr/share/fonts/TTF/XB-Zar-Regular.ttf'
font_prop = FontProperties(fname=font_path)
from matplotlib.animation import FuncAnimation
# We'll use a colormap to assign color to the bars
cmap = plt.get_cmap('tab20')

colors = plt.cm.tab20c.colors  # Colormap with 20 colors

# Plotting
fig, ax = plt.subplots(figsize=(10, 5), dpi=144)

from bidi.algorithm import get_display
import arabic_reshaper

def update(i):
    ax.clear()
    # Sort and get the top 20 for the current date
    top_20 = df_pivot.iloc[i].sort_values(ascending=False).head(20)
    # Plot
    y = np.arange(len(top_20))
    ax.barh(y, top_20.values, color=colors)
    ax.set_yticks(y)
    ylabels = [get_display(arabic_reshaper.reshape(label)) for label in top_20.index]
    ax.set_yticklabels( ylabels)
    date_str = df_pivot.index[i].strftime('%Y-%m-%d')
    ax.set_ylabel('Hashtag', fontproperties=font_prop)
    ax.set_title(f'Top Hashtags on {date_str}')

# Animate
anim = FuncAnimation(fig=fig, func=update, frames=len(df_pivot), interval=500, repeat=False)

# Show animation (in Jupyter Notebook)
from IPython.display import HTML
HTML(anim.to_jshtml())
```
