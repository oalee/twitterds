```python
import pandas as pd

# Step 1: Ensure 'date' column is a datetime object
users['date'] = pd.to_datetime(users['date'])

# Step 2: Create a new column 'month_year' with the month and year information
users['month_year'] = users['date'].apply(lambda x: x.strftime('%Y-%m'))

# Step 3: Group the DataFrame by 'month_year' and aggregate tweets and retweets
grouped_users = users.groupby('month_year').agg({'replyCount': 'sum', 'retweetCount': 'sum', 'likeCount': 'sum', 'quoteCount': 'sum'}).reset_index()

print(grouped_users)
```
